# Question 1
$$\Delta \delta = \delta_{Alcor} - \delta_{Mizar}$$
$$\Delta \alpha = (\alpha_{Alcor} - \alpha_{Mizar}) * 15\ arcsec/sec * \cos(\delta_{Alcor} + \Delta \delta/2)$$
$$d = \sqrt{\Delta \delta^2 + \Delta \alpha^2}$$
$$PA = \tan^{-1} \frac{\Delta \alpha}{\Delta \delta}$$

```python
# For Mizar
mizar = SkyCoord(("13:23:55.54048 +54:55:43.2671"), unit=(u.hourangle, u.deg), frame='icrs')

# For Alcor
alcor = SkyCoord(("13:25:13.53783 +54:59:16.6548"), unit=(u.hourangle, u.deg), frame='icrs')
```

```python
delta_dec = alcor.dec - mizar.dec 
delta_alpha = (alcor.ra - mizar.ra).to(u.arcsec) * np.cos((alcor.dec + delta_dec/2).to(u.rad))
```

```python
d = np.sqrt(delta_alpha**2 + delta_dec**2)
print(d)
```

```python
position_angle = np.arctan2(delta_alpha, delta_dec)
```

```python
position_angle.to(u.deg)
```

The location of WIRO is longitude=$-105^\circ58'33''$ latitude=$+41^\circ 5'49''$. 

From this information, we can then calculate the sky area that is available to observe on Christmas eve. Without considering the airmass, the declination range observable is 
$$41^o \pm 90^o = (-49^o, 90^o)$$
Because the declination cannot go over 90 degrees. 

At Christmas eve, WIRO is about 3 months away from vernal equinox. Thus, the observable objects should have RA within (0h, 12h) range. 
The hour an object stays in observable sky is (dec + 49 - 15) / 90 * 24 = (dec + 34) / 90 * 24 (-15 is to compensate for 15 degree higher than horizon to be observable). It will rise at 6 - (18h - RA). Because the sun is at RA=18h on Christmas eve (Winter solstice).

- The Kitt Peak Direct Imaging Manual: 
The manual describes the available telescopes at the Kitt peak national observatory. In particular, there are four telescope. 4-m telescope, 3.5 m WIYN telescope, 2.1 m and 0.9 m telescope. The document also includes the FOV, limiting magnitude and CCD characteristics on these telescopes. Some setting recommendation for each filter is also provided. 
<br/><br/>
- Photometric Standards: Landolt, 1992, AJ, 104, 340: 
This paper describes the UBVRI Johnson-Kron-Cousins broadband photometric system. The author measured 526 stars with a 298 number set and tries to setup a standard UBVRI photometric filter systems and all other observations relied on.
<br/><br/>
- Photometric Standards: Odewahn et al., 1992, PASP, 104, 553: 
This paper describes the observation result of three galaxies using CCD cameras to setup photometric standards. The authors discuss the error of the standards and derive the photometric systems transformation law. 
<br/><br/>
- Photometric Standards: Christian et al., 1985, PASP, 97, 363: 
The authors describe the photoelectic BVRI photometry of several fields. They try to using these BVRI to setup a standard filters to calibrate detectors. 
<br/><br/>
- JHK Standard stars...: Hawarden et al. 2001: 
This paper presents the JHK filters for Infrared telescope. The standard is put up by UK Infrared Telescope using 82 standard stars. The system is widely accepted by current IR telescopes, such as JWST and Spitzer. 
<br/><br/>
- SpectroPhotometric Standards: Oke, 1990, AJ, 99, 1621: 
This paper describes the setup of spectrophotometric standards using HST and HST related telescope. The standard is based on absolute spectral energy distribution of 25 stars on Hale telescope. 
<br/><br/>
- SpectroPhotometric Standards: Hamuy et al. 1992 PASP, 104, 533
Authors of this paper describe the secondary and tertiary spectrophotometric and UBVRI standard using new observations using Richey-Chretien (RC) specrographs and CCDs. They also discuss the precision of the standard and their result can be useful in spectra calibrations. As the name entails, this standard is designed for southern hemisphere observations.  
<br/><br/>
- SpectroPhotometric Standards: Stritzinger et al. 2005, PASP 117, 810
This paper describes the a six spectrophotometric secondary standard based on 102 Landolt standard observations usign CTIA observatory. This paper aims to provide a standard can be used for both nothern and southern hemisphere when researchers try to calibrate spectra.
<br/><br/>
- ugriz Standards by Smith et al., 2002, AJ, 123, 2121
This paper defines the new ugriz photometric standard, which is mainly designed for SDSS. It is currently one of the most widely used photometric system. The paper also discusses the transformation equation between UBVRI and ugriz system. 
<br/><br/>
- Polarimetric, & Spectrophotometric Standards: Turnshek et al., 1990, AJ, 99, 1243
Because of the versatile capability of HST, this paper presents overview of several standard sources used to calibrate HST photometry, spectrophotomery, polarimetry and images. The authors try to set up standards for calibration of all HST data using these observations. (I'm not sure whether this system is used for different telescopes)
<br/><br/>
- Skyglow: Osterbrock et al., 1996, PASP 108, 277
This paper discuss the atmospheric emission lines by OH, O2 on Keck telescope. This result can be helpful especially when we have very long exposure to get spectrum. This paper tries to identify emission features of some prominent sources in these long exposurs. 
<br/><br/>
- Atmospheric Refraction: Filippenko 1982, PASP, 94, 715
This paper describes the effect of atmosphereic refraction on spectrophotomery. The author shows that the atmospheric refraction changes the observed wavelength of several emission lines. The paper also discussed the relation between object position and optimal slit or aperture orientation to minimize the refractio neffect. 
<br/><br/>
- Bandpasses: Fukugita et al., 1995, PASP, 107, 945
This paper introduces a spectrophotometric synthesis technique to study galaxy colors. The authors compare galaxy colors acuqired by this synthesis method to 
ones publised by Kennicutt and Coleman, Wu, Weeedman in different band passes. The transformation law between different systems is also presented.
<br/><br/>
- Lamp Spectra: CCD Atlas of Helium/Neon/Argon Spectra with lines labeled: E. Carder at Kitt Peak
This artile describes the HeNeAr spectra used for line identification using Gold Camera. This result serves as a starting point in spectroscopy to identify existing emission lines at different gratings. 
<br/><br/>
- Lamp Spectra: CCD Atlas of CuAr Spectra with lines labeled
This article used CuAr lmap to identify several emission lines from the lamp. Similarly to the aricle above, this artcile can also be used to identify emission lines in a spectra, but in different wavelength range.
